```python
#Data acquisition: hexcode and rgb
url = 'https://www.thread-bare.com/tools/closest-dmc-thread-color-lookup'

html = requests.get(url).content
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
#Data aquisition: hexcodes
table = soup.find_all('li', {'class':'svelte-6gtbi1'})
code = re.findall('(?<=data-code=").*(?=style)', str(table))

for i in range(len(code)):
    code[i]= code[i].replace('" ','')
    i+=1
code
```

```python
hex_code = re.findall('(?<=background-color: ).*(?=;")',str(table))
for i in range(len(hex_code)):
    hex_code[i] = hex_code[i].upper()
hex_code
```

```python
df_hex = pd.DataFrame(list(zip(code, hex_code)), columns =['code', 'hexcode'])
df_hex
```

```python
def get_rgb_from_hex(hexcode):
    rgb = matplotlib.colors.to_rgb(hexcode)
    r = int(rgb[0]*255)
    g = int(rgb[1]*255)
    b = int(rgb[2]*255)
    return r,g,b
```

```python
df_hex['r'] = np.nan
df_hex['g'] = np.nan
df_hex['b'] = np.nan

for i in range(df_hex.shape[0]):
    df_hex.loc[df_hex.index==i, 'r']=get_rgb_from_hex(df_hex.iloc[i]['hexcode'])[0]
    df_hex.loc[df_hex.index==i, 'g']=get_rgb_from_hex(df_hex.iloc[i]['hexcode'])[1]
    df_hex.loc[df_hex.index==i, 'b']=get_rgb_from_hex(df_hex.iloc[i]['hexcode'])[2]


df_hex
```

```python
#Fill missing rows with info contained in other rows
def fill_rows_nan(df,columnnan,index,columnfull):
    df.at[index,columnnan]=df[columnfull][index]
    return df[columnnan][index]
```

```python
hexcode_nulls = list(hexcodes[hexcodes['hexcode'].isnull()].index)
for i in range(len(hexcode_nulls)):
    newhexcode = fill_rows_nan(hexcodes,'hexcode',hexcode_nulls[i],'hexcode2')
    newr = fill_rows_nan(hexcodes,'r',hexcode_nulls[i],'r2')
    newg = fill_rows_nan(hexcodes,'g',hexcode_nulls[i],'g2')
    newb = fill_rows_nan(hexcodes,'b',hexcode_nulls[i],'b2')
```

```python
list_hexcode3 = list(hexcodes[hexcodes['hexcode3'].isnull()].index)
for i in range(len(list_hexcode3)):
    updated_hexcode = fill_rows_nan(hexcodes,'hexcode3',list_hexcode3[i],'hexcode')
```
